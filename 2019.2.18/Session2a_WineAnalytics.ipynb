```r
# Read in data
wine = read.csv("wine.csv")
str(wine)
summary(wine)

#Scatter plot between age and price
plot(wine$Age, wine$Price, xlab="Age (Years)", ylab = "Price", col = "red",cex.axis=1.5,cex.lab=1.5,pch=16)

#Scatter plot between age and price
plot(wine$AGST, wine$Price, xlab="Average Growing Season Temperature", ylab = "Price", col = "red",cex.axis=1.5,cex.lab=1.5,pch=16)

#Scatter plot between age and price
plot(wine$HarvestRain, wine$Price, xlab="Harvest Rain", ylab = "Price", col = "red",cex.axis=1.5,cex.lab=1.5,pch=16)

#Scatter plot between age and price
plot(wine$WinterRain, wine$Price, xlab="Winter Rain", ylab = "Price", col = "red",cex.axis=1.5,cex.lab=1.5,pch=16)


# Linear Regression (one variable)
model1 = lm(Price ~ AGST, data=wine)
summary(model1)

#install and load the package 'ggplot2'
#install.packages('ggplot2',repos='https://mirrors.tuna.tsinghua.edu.cn/CRAN/')
library('ggplot2')

#Graphical illustration of Linear Regression
#ggplot(data=wine,aes(x=AGST,y=Price))+geom_point()+geom_smooth(method="lm",se=FALSE)

# Sum of Squared Errors
model1$residuals
SSE = sum(model1$residuals^2)
SSE

# Compute R-squared
SST = sum((wine$Price - mean(wine$Price))^2)
1 - SSE/SST

# Compute the fitted outcomes
predictTrain = predict(model1, newdata=wine)
predictTrain

sum((predictTrain- mean(wine$Price))^2)/SST
```

```r
#Load the wine data
wine = read.csv("wine.csv")
```

```r
#Build the linear regression model with all the covariates
model2 = lm(Price ~ AGST + HarvestRain + WinterRain + Age + FrancePop, data=wine)
summary(model2)
```

```r
#Calculate the correlation between different variables in the data frame wine
cor(wine)
cor(wine$Age,wine$FrancePop)
```

```r
#Build the linear regression model with France population removed from the covariates.
model3 = lm(Price ~ AGST + HarvestRain + WinterRain + Age, data=wine)
summary(model3)
```

```r
##Calculate the out-of-sample R-squared for model3
#Load the testing data
wine_test = read.csv("wine_test.csv")
str(wine_test)

# Make test set predictions
predictTest3 = predict(model3, newdata=wine_test)
predictTest3

# Compute R-squared
SSE3 = sum((wine_test$Price - predictTest3)^2)
SSE3

SST = sum((wine_test$Price - mean(wine$Price))^2)
SST

Rsquared3 = 1 - SSE3/SST
Rsquared3
```

```r
## Price ~ AGST
model1 = lm(Price~AGST,data=wine)
predictTest1 = predict(model1, newdata=wine_test)
predictTest1

# Compute R-squared
SSE1 = sum((wine_test$Price - predictTest1)^2)
SSE1

Rsquared1 = 1 - SSE1/SST
Rsquared1
```

```r
## Price ~ AGST + Harvest Rain
model4 = lm(Price~AGST+HarvestRain,data=wine)
predictTest4 = predict(model4, newdata=wine_test)
predictTest4

# Compute R-squared
SSE4 = sum((wine_test$Price - predictTest4)^2)
SSE4

Rsquared4 = 1 - SSE4/SST
Rsquared4
```

```r
## Price ~ AGST + Harvest Rain + Age
model5 = lm(Price~AGST+HarvestRain+Age,data=wine)
predictTest5 = predict(model5, newdata=wine_test)
predictTest5

# Compute R-squared
SSE5 = sum((wine_test$Price - predictTest5)^2)
SSE5

Rsquared5 = 1 - SSE5/SST
Rsquared5
```

```r
## Price ~ AGST + Harvest Rain + Age + Winter Rain + Population (Model2)
summary(model2)

predictTest2 = predict(model2, newdata=wine_test)
predictTest2

# Compute R-squared
SSE2 = sum((wine_test$Price - predictTest2)^2)
SSE2

Rsquared2 = 1 - SSE2/SST
Rsquared2
```

```r
#Use crossterms to predict child iq
kidiq = read.csv("kidiq.csv")
str(kidiq)
```

```r
# Linear regression model without cross-term
model_linear = lm(kid_score~mom_hs+mom_iq,data=kidiq)
summary(model_linear)
```

For the model without cross-term, we have

kid\_score$\approx$25.73+0.56mom\_iq if mom\_hs=0

kid\_score$\approx$31.68+0.56mom\_iq if mom\_hs=1

```r
#linear regression model with cross-term
model_cross = lm(kid_score~mom_hs+mom_iq+mom_hs*mom_iq,data=kidiq)
summary(model_cross)
```

For the model with cross-term, we have

kid\_score$\approx$-11.48+0.97mom\_iq if mom\_hs=0

kid\_score$\approx$39.79+0.48mom\_iq if mom\_hs=1

```r
##Centering
wine$AGSTc = wine$AGST-mean(wine$AGST)
wine$Agec = wine$Age-mean(wine$Age)
wine$HarvestRainc = wine$HarvestRain-mean(wine$HarvestRain)
wine$WinterRainc = wine$WinterRain-mean(wine$WinterRain)
```

```r
model_c = lm(Price~AGSTc + HarvestRainc+ WinterRainc + Agec,data=wine)
summary(model_c)
summary(model3)
```

```r
#Linear regression with Standardized covariates
wine$AGSTs = wine$AGSTc/sd(wine$AGST)
wine$Ages = wine$Agec/sd(wine$Age)
wine$HarvestRains = wine$HarvestRainc/sd(wine$HarvestRain)
wine$WinterRains = wine$WinterRainc/sd(wine$WinterRain)
```

```r
model_s = lm(Price~AGSTs + HarvestRains+ WinterRains + Ages,data=wine)
```
